```python
df = df[df["ID_VISITE"]==1]
```

```python
now = datetime.now()
df['Date de naissance du patient'] = pd.to_datetime(df['Date de naissance du patient'], format='%m%d%y') 
df['Date de naissance du patient'] = df['Date de naissance du patient'].where(df['Date de naissance du patient'] < now, df['Date de naissance du patient'] -  np.timedelta64(100, 'Y')) 
df['age'] = (now - df['Date de naissance du patient']).astype('<m8[Y]')
```

```python
df = df[df["EXAMEN ENFANTS"] == 'Non']
df_nan = df[df["EXAMEN ENFANTS"].isna()]
df_age = df[df["age"]>=16]
```

```python
#Tableau de Répartition des âges
age = df['age'].value_counts()
age_df = pd.DataFrame(data=age)
age_df.columns = ['Effectif']
age_df.index.name = "Age"
age_df.reset_index(inplace=True)
#Changement de type numérique pour les deux colonnes : float -> int
age_df['Effectif'] = age_df['Effectif'].astype(int)
age_df["Age"] = age_df["Age"].astype(int)
```
